## Simulação de Monte Carlo

- Problema da simulação de Monte Carlo

- Problema da simulação de Monte Carlo utilizando correlação

- Tomada de decisão sob incerteza

- Quantitative finance

### Desafio

```
x1 = | μ1 = 5
     | σ1 = 1

x2 = | μ2 = 1
     | σ2 = 1

      +----------------+
x1 -> | (x1 * x2) ** 3 |
      | ______________ | -> saída = Var95%
x2 -> |      1000      |
      +----------------+
```

````python
import numpy as np
import matplotlib.pyplot as plt

def desafio(x1,x2):
    return ((x1 * x2) ** 3) / 1000

mi1 = 5.0
sigma1 = 1.1

mi2 = 1.0
sigma2 = 1.0

tamanho = 1000000

x1 = np.zeros(tamanho)
x2 = np.zeros(tamanho)

x1 = np.random.normal(mi1,sigma1,tamanho)
x2 = np.random.normal(mi2,sigma2,tamanho)

saida = desafio(x1,x2)

var95perc = np.percentile(saida,95)
var5perc = np.percentile(saida,5)

print('var5 usando percentile: ',var95perc)
print('var95 usando percetile: ',var5perc)

graf = plt.hist(saida,100)
plt.show()
````
