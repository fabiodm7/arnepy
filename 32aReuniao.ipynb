## Random walk

```python
import numpy as np
import matplotlib.pyplot as plt

preco_acao_hoje = 2.0
mu = 0.02 # mean and 
sigma = 0.1 # standard deviation

quant_cenario = 5000
preco_acao_3_semanas = np.zeros((quant_cenario,4))

for cenario in range(quant_cenario):
    preco_acao_3_semanas[cenario,0] = preco_acao_hoje
    return1 = np.random.normal(mu, sigma, 3)
    preco2 = preco_acao_3_semanas[cenario,:]
    for i in range(3):
        preco2[i+1] =  preco2[i] * (1 + return1[i])
    #end for
    plt.plot(preco_acao_3_semanas[cenario])

plt.show()

valor_medio_preco_semana_3 = np.mean(preco_acao_3_semanas[:,3])
desvio_padrao_semana_3 = np.std(preco_acao_3_semanas[:,3])

print("valor_medio_preco_semana_3: {:.2f}".format(valor_medio_preco_semana_3))
print("desvio_padrao_semana_3: {:.5f}".format(desvio_padrao_semana_3))

```
